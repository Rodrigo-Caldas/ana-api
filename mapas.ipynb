```python
import geopandas as gpd
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

## Bacia do Grande

```python
shp_brasil = 'brasil/BR_UF_2022.shp'
shp_grande = 'bacias-sem-sub-bacias/grande.shp'
shp_agua_vermelha = 'sub-bacias-isoladas/grande/agua_vermelha.shp'

gdf_brasil = gpd.read_file(shp_brasil)
gdf_grande = gpd.read_file(shp_grande)
gdf_agua_vermelha= gpd.read_file(shp_agua_vermelha)
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

gdf_brasil.plot(ax=ax, color='orange', edgecolor='k', alpha=0.5)
gdf_grande.plot(ax=ax, color='green', edgecolor='None', alpha=0.9,)
gdf_agua_vermelha.plot(ax=ax, color='yellow', edgecolor='k', alpha=0.5)

# Ajustar os limites do gr√°fico para a bounding box do shapefile menor
ax.set_xlim([-51, -43])  # Definir os limites no eixo X
ax.set_ylim([-24, -18])  # Definir os limites no eixo Y

# Criar legendas manualmente
handles = [
    patches.Patch(color='orange', edgecolor='k', alpha=0.5, label='Brasil'),
    patches.Patch(color='green', edgecolor='None', alpha=0.9, label='Bacia do Rio Grande'),
    patches.Patch(color='green', edgecolor='k', alpha=0.55, label='Sub-bacia √Ågua Vermelha')
]

# Mostrar o gr√°fico
plt.title('Sub-bacia √Ågua Vermelha', fontsize=18, loc='center',)
ax.legend(handles=handles, loc='upper right')
plt.show()
```
