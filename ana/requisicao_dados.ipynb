```python
import ana
import utils
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
shp_agua_vermelha = gpd.read_file("../Mapas/sub-bacias-isoladas/grande/agua_vermelha.shp")

df_inventario = ana.mostrar_inventario()
gdf_inventario_filtrado = utils.filtrar_postos_da_bacia(df_inventario, shp_agua_vermelha)
```

```python
gdf_codigos = gdf_inventario_filtrado[gdf_inventario_filtrado["operando"] == "sim"]
lista_codigo = gdf_codigos["codigo"].tolist()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

shp_agua_vermelha.plot(ax=ax, color='orange', edgecolor='k', alpha=0.5)
gdf_codigos.plot(ax=ax, color='red', marker='o', markersize=20)
```
